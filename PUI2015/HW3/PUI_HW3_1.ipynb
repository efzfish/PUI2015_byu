## Sample Plot

```python
mysize=1000/(np.array(range(1,100)+[10]))
md['pois'] = {}
md['gauss'] = {}
md['chisq'] = {}
md['geo'] = {}
md['exp'] = {}
md['bin'] = {}
for n in mysize:
    md['pois'][n] = np.random.poisson(100, size=n)
    md['gauss'][n] = np.random.normal(100, 100, size=n)
    md['chisq'][n] = np.random.chisquare(100, size=n)
    md['geo'][n] = np.random.geometric(0.01, size=n)
    md['exp'][n] = np.random.exponential(100, size=n)
    md['bin'][n] = np.random.binomial(400, 0.25, size=n)

fig_mu_N = pl.figure(figsize=(24,16))
gs = gridspec.GridSpec(2, 3)
means = {}

means['pois'] = {}
axchisq_mu_n = fig_mu_N.add_subplot(gs[0,0])
for nn in md['pois'].iterkeys():
    means['pois'][nn] = md['pois'][nn].mean()
    axchisq_mu_n.plot(nn, means['pois'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')

means['gauss'] = {}
axchisq_mu_n = fig_mu_N.add_subplot(gs[0,1])
for nn in md['gauss'].iterkeys():
    means['gauss'][nn] = md['gauss'][nn].mean()
    axchisq_mu_n.plot(nn, means['gauss'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')

means['chisq'] = {}
axchisq_mu_n = fig_mu_N.add_subplot(gs[0,2])
for nn in md['chisq'].iterkeys():
    means['chisq'][nn] = md['chisq'][nn].mean()
    axchisq_mu_n.plot(nn, means['chisq'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')
    
means['geo'] = {}
axchisq_mu_n = fig_mu_N.add_subplot(gs[1,0])
for nn in md['geo'].iterkeys():
    means['geo'][nn] = md['geo'][nn].mean()
    axchisq_mu_n.plot(nn, means['geo'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')
    
means['exp'] = {}
axchisq_mu_n = fig_mu_N.add_subplot(gs[1,1])
for nn in md['exp'].iterkeys():
    means['exp'][nn] = md['exp'][nn].mean()
    axchisq_mu_n.plot(nn, means['exp'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')
    
means['bin'] = {}
axchisq_mu_n = fig_mu_N.add_subplot(gs[1,2])
for nn in md['bin'].iterkeys():
    means['bin'][nn] = md['bin'][nn].mean()
    axchisq_mu_n.plot(nn, means['bin'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')
    
```
